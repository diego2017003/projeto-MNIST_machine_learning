```python
import numpy as np
import tensorflow_datasets as tfds
import tensorflow
import keras
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()


x_train
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,20))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i].reshape((28,28)), cmap = 'gray')
```

```python
import pandas as pd
import seaborn as sns
train_vals_count = pd.Series(y_train).value_counts()
test_vals_count = pd.Series(y_test).value_counts()
plt.figure(figsize=(14,14))
plt.subplot(2,1,1)
sns.barplot(x=train_vals_count.keys(),y=train_vals_count.values).set_title("train labels ocurrencies")
plt.subplot(2,1,2)
sns.barplot(x=test_vals_count.keys(),y=test_vals_count.values).set_title("test labels ocurrencies")
```
